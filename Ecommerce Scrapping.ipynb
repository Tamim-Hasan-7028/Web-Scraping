Lets scrap data from StarTech. The goal is to find ANC TWS of Edifier, Realme and QYC from startech which is in stock and sort the price from low to high. After filtering, we have to scrap the name, details and updated price of TWS from the result and store it as csv format.

(This work was done in 8 July, 2024. Page content may get updated in current time. So code may vary and not work properly)

```python
#import library

from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
driver = webdriver.Chrome()

#entering website
url = 'https://www.startech.com.bd/'
driver.get(url)
# driver.maximize_window()
time.sleep(3)

#selecting "earbuds" button
driver.find_element(By.XPATH,'/html/body/div[5]/div[1]/div[4]/div/div[14]/a').click()
time.sleep(2)

#selecting "sort"
driver.find_element(By.XPATH,'//*[@id="input-sort"]').click()
time.sleep(2)

#selecting "low to high"
driver.find_element(By.XPATH,'//*[@id="input-sort"]/option[2]').click()
time.sleep(2)

#selecting "filter"
driver.find_element(By.XPATH,'//*[@id="lc-toggle"]').click()
time.sleep(2)

#selecting "in stock"
driver.find_element(By.XPATH,'//*[@id="column-left"]/div/div[2]/div[2]/label[1]/span').click()
time.sleep(2)

#selecting "edifier"
driver.find_element(By.XPATH,'//*[@id="column-left"]/div/div[3]/div[2]/label[4]/span').click()
time.sleep(2)

#selecting "realme"
driver.find_element(By.XPATH,'//*[@id="column-left"]/div/div[3]/div[2]/label[5]/span').click()
time.sleep(2)

#selecting "QYC"
driver.find_element(By.XPATH,'//*[@id="column-left"]/div/div[3]/div[2]/label[16]/span').click()
time.sleep(2)

#selecting "ANC"
driver.find_element(By.XPATH,'//*[@id="column-left"]/div/div[5]/div[2]/label[1]/span').click()
time.sleep(2)

#selecting "cross button"
driver.find_element(By.XPATH,'//*[@id="column-left"]/span/i').click()
time.sleep(2)
```

```python
#creating list
product_names = []
product_details = []
product_prices = []

#scraping 16 earbuds data after filtering
for i in range(1, 17):  # Loop from 1 to 16
    # Construct the XPath for each product
    product_xpath = f'//*[@id="content"]/div[2]/div[{i}]'

    name = driver.find_element(By.XPATH, f'{product_xpath}/div//h4').text  # Adjust as needed
    product_names.append(name)

    details = driver.find_element(By.XPATH, f'{product_xpath}/div//div[@class="short-description"]').text
    product_details.append(details)
    
    price = driver.find_element(By.XPATH, f'{product_xpath}/div//div[@class="p-item-price"]/span[1]').text  # Adjust as needed
    product_prices.append(price)

#separating data by comma
product_details_comma_separated = [detail.replace('\n', ', ') for detail in product_details]

# print(product_names)
# print(product_prices)
# print(product_details_comma_separated)


## creating x-path
#product_xpath = //*[@id="content"]/div[2]/div[{i}]
# name= ./div//h4
# description= ./div//div[@class="short-description"]
# price= ./div//div[@class="p-item-price"]/span[1]
```

```python
# Convert the data to a DataFrame
df = pd.DataFrame({
    'Product Name': product_names,
    'Product Details': product_details_comma_separated,
    'Price': product_prices
})

print(df)

# Save the DataFrame to a CSV file
df.to_csv('C:/Users/Saiful Hasan/Downloads/StarTech-ANC-tws.csv', index=False)

# Close the browser
driver.quit()
```
